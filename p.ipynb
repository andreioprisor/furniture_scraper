```python
with open("websites.txt", "r") as f:
	urls = [line.split(",")[0][1:-1] for line in f.readlines()]
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from webdriver_manager.chrome import ChromeDriverManager
import time
from bs4 import BeautifulSoup


# Setting up the Selenium WebDriver
options = Options()
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=options)

# Function to extract text from a website
for url in urls:
	driver.get(url)
	driver.implicitly_wait(2)
	try:
		a = driver.find_element(By.XPATH, """//*[contains(translate(string(@*),'ACCEPT','accept'),'accept') or 
                         			contains(translate(string(@*), 'CLOSE', 'close'), 'close')]""")
		a.click()
	except:
		pass
	body_text = driver.page_source
	time.sleep(1)
	soup = BeautifulSoup(body_text, 'html.parser')
	a = soup.get_text(separator=" | ",strip=True).split(" | ")
	with open ("text.txt", "a") as f:
		for i in a:
			f.write(f"{i}\tfalse\n")
```
